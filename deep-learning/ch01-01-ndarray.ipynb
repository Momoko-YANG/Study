1. 张量的创建

ones函数创建一个具有指定形状的新张量，并将所有元素值设置为1

```python
t = torch.ones(4)
print('t:', t)
```

```python
x = torch.arrange(12)
print('x:', x)
print('x shape:', x.shape) # 访问向量的形状
```

```python
y = x.reshape(3, 4)  # 改变一个张量的形状而不改变元素数量和元素值
print('y:', y)
print('y.numel():', y.numel())  # 返回张量中元素的总个数
```

```python
z = torch.zeros(2, 3, 4)  # 创建一个张量，其中所有元素都设置为0
print('z:', z)
```

```python
w = torch.randn(2, 3, 4) # 每个元素都从均值为0、标准差为1的标准高斯（正态）分布中随机采样。
print('w:', w)
```

通过提供包含数值的 Python 列表（或嵌套列表）来为所需张量中的每个元素赋予确定值

```python
q = torch.tensor([[1, 2, 3], [4, 3, 2], [7, 4, 3]]) 
print('q:', q)
```

2. 张量的运算

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x ** y)
print(torch.exp(x))
```

```python
X = torch.arange(12, dtype = torch.float32).reshape(3, 4)
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print('cat操作 dim=0', torch.cat((x, Y), dim=0))
print('cat操作 dim=1', torch.cat((X, Y), dim=1))  # 连结（concatenate） ,将它们端到端堆叠以形成更大的张量。
```

```python
print('X == Y', X == Y) # 通过 逻辑运算符 构建二元张量
print('X < Y', X < Y)
print('张量所有元素的和:', X.sum())
```

3. 广播机制

```python
a = torch.arange(3).reshape(3, 1)
b = torch.arange(2).reshape(1, 2)
print('a:', a)
print('b:', b)
print('a + b:', a + b)
```

4. 索引和切片

```python
X = torch.arange(12, dtype=torch.float32).reshape(3, 4)
print('X:', X)
print('X[-1]:', X[-1]) # 用 [-1] 选择最后一个元素
print('X[1:3]:', X[1:3]) # 用 [1:3] 选择第二个和第三个元素]
```

```python
X[1, 2] = 9 # 写入元素
print('X:', X)
```

```python
X[0:2, :] = 12 # 写入元素
print('X:', X)
```

5. 节约内存

```python
before = id(Y) # id()函数提供了内存中引用对象的确切地址
Y = Y + X
print(id(Y) == before)
```

```python
before = id(X)
X += Y
print(id(X) == before)  # 使用 X[:] = X + Y 或 X += Y 来减少操作的内存开销。
```

```python
before = id(X)
X[:] = X + Y
print(id(X) == before)  # 使用 X[:] = X + Y 或 X += Y 来减少操作的内存开销。
```
